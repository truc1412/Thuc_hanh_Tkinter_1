yêu cầu người dùng nhập

 message “Hello” và
tên của họ và thay đổi

### Bài 3

thành tổng số và hiển thị nó trong một
button khác. Điều này có thể là
lặp đi lặp lại nhiều lần như
họ muốn và tiếp tục thêm vào
Tổng số. Nên có một button
khác thiết lập lại
tổng trở về 0 và làm trống
bản gốc text box , sẵn sàng cho
họ để bắt đầu lại.

### Bài 4
Tạo một cửa sổ sẽ yêu cầu người dùng nhập một
tên trong một text box . Khi họ bấm vào một button nó
sẽ thêm nó vào cuối danh sách được hiển thị trên
màn hình. Tạo một button khác sẽ xóa
danh sách.

```python
from tkinter import *

window=Tk()
window.geometry("400x250")
window.title("Name")

def add():
    global name,nameVar
    name.append(entry.get())
    nameVar.set(name)
    entry.delete(0,END)
    
def clear():
    global name,nameVar
    name=[]
    nameVar.set(name)
    entry.delete(0,END)
    entry.focus()

label= Label(window,text="Enter a name:")
entry = Entry(window,width=20)

b1= Button(window,text="Add", padx=13, pady=10, command=add)
b2= Button(window,text="Clear", padx=11, pady=10, command=add)

name=[]
nameVar=StringVar(value=name)
lst = Listbox(window, listvariable=nameVar, height=15)

label.place(x=47, y=30)
entry.place(x=20, y=60)
lst.place(x=160, y=30)
b1.place(x=50, y=110)
b2.place(x=50, y=180)

window.mainloop()
```

### Bài 5
1 km = 0,6214 dặm và 1
dặm = 1,6093 km. sử dụng
những con số này, làm cho một chương trình
điều đó sẽ cho phép người dùng
chuyển đổi giữa dặm và
km.

```python
from tkinter import *

def dam_to_km():
    dam = dam_chuyen.get()
    km =float(dam)* 1.6093
    result_km["text"] = f"{round(km, 4)} km"

def km_to_dam ():
    km = km_chuyen.get()
    dam = float(km) * 0.6214
    result_dam["text"] = f"{round(dam, 4)} dam"


window = Tk()
window.title("Converter")
window.resizable(width=False, height=False)

nhap_dam = Frame(master=window)
dam_chuyen= Entry(master=window, width=10)
lbl_dam = Label(master=window, text="DAM")

dam_chuyen.grid(row=0, column=0, sticky="e")
lbl_dam.grid(row=0, column=1, sticky="w")

btn_convert_dam = Button(
    master=window,
    text="\N{RIGHTWARDS BLACK ARROW}",
    command=dam_to_km
)

result_km = Label(master=window, text="Km")

nhap_dam.grid(row=0, column=0, padx=10)
btn_convert_dam.grid(row=0, column=2, pady=10)
result_km.grid(row=0, column=3, padx=10)


nhap_km = Frame(master=window)
km_chuyen = Entry(master=window, width=10)
lbl_km = Label(master=window, text="KM")

km_chuyen.grid(row=1, column=0, sticky="e")
lbl_km.grid(row=1, column=1, sticky="w")

btn_convert_km = Button(
    master=window,
    text="\N{RIGHTWARDS BLACK ARROW}",
    command=km_to_dam
)

result_dam = Label(master=window, text="DAM")

nhap_km.grid(row=1, column=0, padx=10)
btn_convert_km.grid(row=1, column=2, pady=10)
result_dam.grid(row=1, column=3, padx=10)



window.mainloop()
```
